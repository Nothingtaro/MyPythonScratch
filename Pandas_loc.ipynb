## Pandas: loc_1

```python
import pandas as pd
pd.__version__
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mtcars.csv',
                nrows=10, index_col='model')
df
```

### single label 

```python
df.loc['Datsun 710']
```

```python
df.loc[['Datsun 710']]
```

```python
df.loc[['Datsun 710'], :]
```

```python
df.loc[['Datsun 710'], 'mpg']
```

### list of labels 

```python
df.loc[['Datsun 710', 'Merc 230']]
```

```python
df.loc[['Datsun 710', 'Merc 230'], 'mpg':'hp']
```

```python
df.loc[['Datsun 710', 'Merc 230'], ['mpg','hp']]
```

### slice object 

```python
df.loc['Datsun 710':'Merc 230'] # inclusive
```

```python
df.loc['Datsun 710':'Merc 230', 'hp':'gear'] 
```

### boolean array 

```python
df.index
```

```python
df.index.str.contains('Merc')
```

```python
df.loc[df.index.str.contains('Merc')]
```

```python
df.loc[df.index.str.contains('Merc'), 'mpg']
```

```python
df.loc[df.index.str.contains('Merc'), 'mpg':'hp']
```

## Pandas: loc_2

### .loc - select column technique 

```python
import pandas as pd
```

```python
df = pd.read_table('https://github.com/prasertcbs/tutorial/raw/master/score_a.tsv'
                  , index_col='studentid')
df
```

```python
df.columns
```

```python
df.loc[:, df.columns.str.contains('eng')]
```

```python
df.loc[:, df.columns.str.startswith('eng')]
```

```python
df.loc[:, df.columns.str.contains('2')]
```

```python
df.loc[:, df.columns.str.endswith('2')]
```

```python
df.loc[:, df.columns.str.contains('eng|sci')]
```

```python
df.index
```

```python
df.index.str.startswith('B')
```

```python
df.loc[df.index.str.startswith('B')]
```

```python
df.loc[df.index.str.startswith('B'), df.columns.str.endswith('3')]
```

```python
df.loc[df.index.str.startswith('B'), df.columns.str.endswith('3')].sum()
```

```python
df.loc[df.index.str.startswith('B'), df.columns.str.endswith('3')].sum(axis=1)
```

```python
df.loc[df.index.str.startswith('B'), df.columns.str.endswith('3')].mean()
```

```python
df.loc[df.index.str.startswith('B'), df.columns.str.endswith('3')].mean(axis=1)
```

```python
df.columns.str.match('(math|sci)[1|3]')
```

```python
df.loc[:, df.columns.str.match('(math|sci)[1|3]')]
```
