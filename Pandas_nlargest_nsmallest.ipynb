## Rows with largest/smallest column values 

```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.head()
```

```python
df.shape
```

```python
df.nlargest(3, columns='cty')
```

```python
df.nlargest(1, columns='cty')
```

```python
df.nsmallest(3, columns='cty')
```

```python
df.trans.str[0]
```

```python
df['trans_am'] = df.trans.str[0]
df.sample(5)
```

```python
df[df.trans_am == 'a'].nlargest(3, columns = 'cty')
```

```python
df[df.trans_am == 'm'].nlargest(3, columns = 'cty')
```

```python
g = df.groupby('trans_am')
g.groups
```

```python
g.get_group('a')
```

```python
g.max()
```

## split-apply-combine 

```python
g.get_group('a').nlargest(3, columns='cty')
```

```python
g.get_group('m').nlargest(3, 'cty')
```

```python
# g.nlargest(3, 'cty') # error
```

```python
g.apply(lambda grp: grp.nlargest(3, 'cty'))
# df.groupby('trans_am').apply(lambda grp: grp.nlargest(3, 'cty'))
```

```python
df.groupby('class').apply(lambda grp: grp.nlargest(1, 'cty'))
```

```python
df.groupby('class').apply(lambda grp: grp.nlargest(1, 'cty')).sort_values('cty', ascending=False)
```
