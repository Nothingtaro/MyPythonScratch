# Example 2: qnaire
Source: https://github.com/prasertcbs/tutorial/raw/master/qnaire_missing_values.csv

```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/qnaire_missing_values.csv' , index_col = 'obs')
df
```

```python
df.dropna()
```

```python
df.shape
# 15 rows , 12 columns
```

```python
df.dropna(thresh=10)
```

```python
df.columns
```

```python
df.dropna(thresh=8, subset=['eq1', 'eq2', 'eq3', 'eq4', 'eq5', 'eq6', 'eq7', 'eq8',
       'eq9', 'eq10'])
```

```python
df.columns.str.contains('eq')
```

```python
df.columns[df.columns.str.contains('eq')]
```

```python
df.dropna(thresh=8, subset=df.columns[df.columns.str.contains('eq')])
```

```python
df_clean = df.dropna(thresh=10, subset=df.columns[df.columns.str.contains('eq')])
df_clean
```

```python
df_clean.mean()
```

# Fill missing values
Source: https://github.com/prasertcbs/tutorial/raw/master/qnaire_fill_missing.tsv

```python
import pandas as pd
```

```python
df = pd.read_table('https://github.com/prasertcbs/tutorial/raw/master/qnaire_fill_missing.tsv'
                  , index_col = 'obs', parse_dates=['dob'])
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.salary.isnull().sum()
```

```python
df
```

```python
df.describe()
```

```python
df.salary.fillna(100500, inplace=True)
```

```python
df
```

```python
df.own_car.value_counts()
```

```python
df.own_car.fillna(False, inplace=True)
```

```python
df
```

```python
df.fillna(df.mean()['csi1':'q3'], inplace=True)
```

```python
df
```

```python
df.dob.fillna(pd.to_datetime('1995-04-13'), inplace=True)
df
```

```python
df.gender.fillna('missing', inplace=True)
df
```
