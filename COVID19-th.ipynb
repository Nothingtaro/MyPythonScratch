```python
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
url = 'https://www.th-stat.com/json/covid19v2/getTimeline.json'
r = requests.get(url)
r.text
```

```python
j = json.loads(r.text)
j
```

```python
j['Data']
```

```python
len(j['Data'])
```

```python
df = pd.DataFrame(j['Data'])
df
```

```python
df.set_index('Date', inplace=True)
df
```

```python
df.info()
```

```python
pd.Timestamp.now()
```

```python
dfilter = df[df['Date'].between('2021-04-01', pd.Timestamp.now())]
dfilter
```

```python
g = sns.relplot(kind='line', x="Date", y="Confirmed", data=dfilter, 
                color='salmon',
                facet_kws={'sharey': True, 'sharex': True})
g;
```
