```python
# first solution
def remove_element(lst: list, s:str):
    try:
        lst.remove(s)
    except:
        pass
```

```python
df.apply(lambda r: remove_element(r['tags'], ('attributed-no-source')), axis=1)
```

```python
# second solution
df.tags.apply(lambda r: [x for x in r if x != 'attributed-no-source'])
```

```python
df['tags'] = df.tags.apply(lambda r: [x for x in r if x != 'attributed-no-source'])
```
