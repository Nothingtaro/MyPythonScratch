## pandas.set_option
doc: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.set_option.html

```python
import pandas as pd
pd.set_option('display.max_rows', None, # None = show all rows
              'display.max_columns', 100, 
              'display.float_format', '{:,.0f}'.format)
```

```python
print(f'pandas version: {pd.__version__}')
```
