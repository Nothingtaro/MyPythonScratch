### String methods

```python
df[df.manufacturer.str.startswith('f')] # start with doesn't support regex
```

```python
df[df.manufacturer.str.match('[af]')] # start with 'a' or 'f'
```

```python
df[df.manufacturer.str.match('[^adf]')] # not start with 'a' or 'd' or 'f'
```

```python
df[df.manufacturer.str.match('[a-f]')] # start with 'a' to 'f' (abcdef)
```

```python
df[df.model.str.match('a\d')] # start with 'a' and follow by number
```

```python
df[df.model.str.match('\d')] # start with number
```

```python
df[df.model.str.match('.i')] # ตัวที่ 2 เป็นตัว i
```
