### read ALL sheets

```python
url = 'https://github.com/prasertcbs/basic-dataset/raw/master/PisaScore2015.xlsx'
df = pd.read_excel(url, sheet_name=None)
```

```python
pisa = pd.merge(df['math'], df['sci'], on='Country')
pisa.head(10)
```

```python
pisa = pd.merge(df['math'][['Country', 'Maths']], df['sci'][['Country', 'Science']], on='Country')
pisa.head(10)
```

```python
pisa = pd.merge(pisa, df['reading'][['Country', 'Reading']], on='Country')
pisa.head(10)
```

```python
pisa['Total'] = pisa.Maths + pisa.Science + pisa.Reading
pisa.head()
```

```python
pisa['Rank'] = pisa['Total'].rank(ascending=False, method='min').astype(int)
pisa
```

```python
pisa.sort_values('Rank').head(10)
```

```python
pisa.to_csv('pisa2015.csv', index=False)
```
