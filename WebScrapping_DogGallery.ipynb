```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

pd.set_option('display.max_colwidth', -1)
```

```python
r = requests.get('https://www.pedigree.com.au/dog-breed-information/dog-breed-gallery/')
```

```python
r.text[:200]
```

```python
s = BeautifulSoup(r.text, 'lxml')
s
```

```python
type(s)
```

```python
# d = s.find('div' , {'id' : 'dogAZ'})
d = s.find('div', id='dogAZ')
d
```

```python
# a_tag = d.find_all('a')
a_tag = d('a')
a_tag
```

```python
# image
a_tag[3].find('img')['src']
```

```python
rows = []
url_prefix = 'https://www.pedigree.com.au/'
for e in a_tag:
    img = e.find('img')
    breed = img['alt']
    src = url_prefix + img['src']
    img_html = f'<img src="{src}"/>'
    link = url_prefix + e['href']
    href_html = f'<a href="{link}">{breed}</a>'
    rows.append((breed, link, src, href_html, img_html))
```

```python
rows
```

```python
df = pd.DataFrame(rows, columns=['breed', 'link', 'img_url', 'href_html', 'img_html'])
df
```

```python
df
```

```python
df.to_csv('dogs.csv', index=False)
```
