```python
dt = df['text'].str.extract(r'(\w+)\s+\(([\u0E00-\u0E7F]+)\)\s+\=\s+([\u0E00-\u0E7F]+)')
dt.columns = ['en', 'pron', 'th']
dt
```

```python
dt = df['text'].str.extract(r'(?P<eng>\w+)\s+\((?P<pron>[\u0E00-\u0E7F]+)\)\s+\=\s+(?P<thai>[\u0E00-\u0E7F]+)')
# dt.columns = ['en', 'pron', 'th']
dt
```

```python
dt = df['text'].str.extract(r'(\w+)\s+\((\D+)\)\s+\=\s+(\D+)')
dt.columns = ['en', 'pron', 'th']
dt
```
