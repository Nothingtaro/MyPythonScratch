## three-level index

```python
df3 = pd.read_table('https://github.com/jennybc/gapminder/raw/master/inst/extdata/gapminder.tsv',
                   index_col = ['continent', 'country', 'year'])
df3
```

```python
df3.sort_index(inplace=True)
df3
```

```python
df3.index
```

```python
df3.index.names
```

```python
df3.index.levels
```

```python
df3.index.nlevels
```

```python
df3.loc['Asia']
```

```python
df3.loc[('Asia', 'Thailand')]
```

```python
df3.loc[('Asia', 'Thailand')].loc[1972:2002]
```

```python
df3.loc[('Asia', 'Thailand')].loc[1972:2002]['pop']
```

```python
df3.loc[('Asia', 'Thailand')].loc[1972:2002][['pop']]
```

```python
%matplotlib inline
df3.loc[('Asia', 'Thailand')].loc[1972:2002]['pop'].plot();
```

```python
df3.index.get_level_values(0)
```

```python
df3.index.get_level_values(1)
```

```python
df3.index.get_level_values(2)
```

```python
df3[df3.index.get_level_values(2) == 2002]
```
