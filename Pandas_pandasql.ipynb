```python
import pandas as pd
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
url = 'https://github.com/prasertcbs/basic-dataset/raw/master/imdb_small/movies.csv'
dmovie = pd.read_csv(url)
dmovie.head()
```

```python
dq = pysqldf("SELECT * FROM dmovie ORDER BY rank DESC")
dq
```

```python
dq = pysqldf("SELECT name, year FROM dmovie ORDER BY rank DESC")
dq
```

```python
dq = pysqldf("SELECT * FROM dmovie ORDER BY rank DESC")
dq
```

```python
dq = pysqldf("SELECT year, COUNT(*) AS freq FROM dmovie WHERE year > 2000 GROUP BY year")
dq
```

```python
sql_cmd = '''
    SELECT year, COUNT(*) AS freq 
    FROM dmovie 
    WHERE year > 2000 
    GROUP BY year
'''

dx = pysqldf(sql_cmd)
dx
```

```python
url = 'https://github.com/prasertcbs/basic-dataset/raw/master/imdb_small/movies_genres.csv'
dgenre = pd.read_csv(url)
dgenre.head()
```

```python
dq = pysqldf("SELECT distinct genre FROM dgenre")
dq
```

```python
sql_cmd = '''
    SELECT * 
    FROM dmovie m INNER JOIN dgenre g ON m.id = g.movie_id
    WHERE genre = 'Action'
'''

dq = pysqldf(sql_cmd)
dq
```
