##  Method 1: Windows authentication

```python
driver = 'ODBC Driver 17 for SQL Server'
server = '.\sqlexpress'
database = 'Yummi2012'

# use trusted connection (Windows authentication)
con_string = f'DRIVER={driver};SERVER={server};DATABASE={database};trusted_connection=yes;'
con = pyodbc.connect(con_string)
```

## Method 2: use SQL Server authentication 

```python
driver = 'ODBC Driver 17 for SQL Server'
server = '.\sqlexpress'
database = 'Yummi2012'

uid = 'tar'
pwd = 'z1x2c3v4'

con_string = f'DRIVER={driver};SERVER={server};DATABASE={database};UID={uid};PWD={pwd}'
print(con_string)
con = pyodbc.connect(con_string)
```

### use Yummi2012 to display columns containing Thai UTF-8

```python
# get data from table
sql_cmd = """
    select MenuId, Descr, DescrTh, Price
        from Menu
        order by price desc
"""
df = pd.read_sql(sql_cmd, con)
df
```

```python
# get data from two tables
sql_cmd = """
    select MenuId, m.Descr, DescrTh, Price, c.descr
        from Menu m
        inner join category c on m.categoryid = c.categoryid
"""
df = pd.read_sql(sql_cmd, con)
df
```
