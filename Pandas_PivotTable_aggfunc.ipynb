```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/long2.csv')
df.head()
```

```python
# df.to_excel('pivot_table.xlsx', index=False)
```

```python
df.pivot_table(index='branch', columns='menu', values='units') # default setting (aggfunc) = mean
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc='sum')
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc='sum', 
              margins=True)
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc='sum', 
              margins=True, margins_name='Grand Total')
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc='sum', 
              margins=True, margins_name='Grand Total')
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc=['sum', 'mean'], 
              margins=True, margins_name='Grand Total')
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc=[np.sum, np.mean], 
              margins=True, margins_name='Grand Total')
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc=[np.sum, np.mean], 
              margins=True, margins_name='Grand Total').round(1)
```

```python
df.pivot_table(index=['branch', 'period'], columns='menu', values='units', margins=True).round(2)
```

```python
df.pivot_table(index='branch', columns=['menu', 'period'], values='units', margins=True).round(2)
```

```python
df.pivot_table(index='branch', columns=['menu', 'period'], values='units', margins=True).round(2).plot(kind='barh');
```

```python
df.pivot_table(index='branch', columns='menu', values='units', margins=True, aggfunc=np.sum).round(2).plot(kind='barh');
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc=np.sum).round(2).plot(kind='barh');
```
