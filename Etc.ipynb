# Find GPA for a student based on letter grades 

```python
print('Welcome to the GPA calculator.')
print('Please enter all your letter grades, one per line.')
print('Enter a blank line to designate the end.')
# map from letter grade to point value
points = { 'A+' : 4.0, 'A' : 4.0, 'A-' : 3.67, 'B+' : 3.33, 'B' : 3.0, 'B-' : 2.67,
           'C+' : 2.33, 'C' : 2.0, 'C' : 1.67, 'D+' : 1.33, 'D' : 1.0, 'F'  : 0.0 }
num_courses = 0
total_points = 0
done = False
while not done:
    grade = input() # read line from user
    if grade == '' : # empty line was entered
        done = True
    elif grade not in points: # unrecognized grade entered
        print("Unknown grade {0} being ignored".format(grade))
    else:
        num_courses += 1
        total_points += points[grade]
if num_courses > 0: # avoid division by zero
    print('Your GPA is {0:.3}'.format(total_points / num_courses))
```

```python
def compute_gpa(grades, points = { 'A+' : 4.0, 'A' : 4.0, 'A-' : 3.67, 'B+' : 3.33, 'B' : 3.0, 'B-' : 2.67,
           'C+' : 2.33, 'C' : 2.0, 'C' : 1.67, 'D+' : 1.33, 'D' : 1.0, 'F'  : 0.0 }):
    num_courses = 0
    total_points = 0
    for g in grades:
        if g in points: # a recognizable grade
            num_courses += 1
            total_points += points[g]
    return total_points / num_courses

compute_gpa('A')
```

# Object Oriented Program (OOP) 

```python
class Eyeglasses:
    def __init__(self, eye, bridge, temple):
        self.eye = eye
        self.bridge = bridge
        self.temple = temple
        
    @classmethod # function decorator
    def of(cls, frame_string): # constructor
        s = frame_string.split("-")
        return cls( int(s[0]), int(s[1]), int(s[2]) )
        
    def __str__(self):
        return "{}-{}-{}".format(self.eye, self.bridge, self.temple)
    
# if __name__ == '__main__':
f1 = Eyeglasses(55, 16, 140)
print(f1)
f2 = Eyeglasses.of("55-16-140")
print(f2)
print(f2.eye, f2.bridge, f2.temple)
```

# Find and sum all integers in the list

```python
import re
def FindInteger(astring):
    numbers = []
    for word in re.findall('\d+', astring):
            numbers.append(int(word))
    return numbers

def SumInteger(astring):
    total = 0
    for j in FindInteger(astring):
        total += j
    return total
                 
astring = input('please enter any string ---> ')
Find = FindInteger(astring)
TotalSum = SumInteger(astring)
print(Find)
print(f'Sum of all elements in given list: {TotalSum}')
```

# Prime Number 

```python
def PrimeNumber(nums):
    pos = -1
    if nums > 1:
        for k in range(2, nums):
            if nums % k == 0:
                pos = 1
        if pos:
            print('Prime')
        else:
            print('Not Prime')
    else:
        print('Not Prime')
    return nums

PrimeInput = int(input('Enter an integer -> '))
print(PrimeNumber(PrimeInput))
```

output: [0,3] (return index)

```python
def TwoSum(arr, target):
    values = {}
    for i, num in enumerate(arr):
        comp = target - num
        if comp in values:
            return([values[comp], i])
        elif comp not in values:
            values[num] = i
        else:
            return []
```

```python
TwoSum([5, 8, 7, 9, 4], 12)
```
