```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.head()
```

```python
sns.boxplot(data=df['cty']);
```

```python
sns.boxplot(data=df[['cty']]);
```

```python
sns.boxplot(data=df[['cty', 'hwy']]);
```

```python
sns.boxplot(x='drv', y='cty', data=df);
```

```python
sns.boxplot(x='cty', y='drv', data=df);
```

```python
df['drv'] = df['drv'].map({'f':'front', 'r':'rear', '4':'4-wheel'}).astype("category")
df
```

```python
sns.boxplot(x='cty', y='drv', data=df, order=['front', '4-wheel', 'rear']);
```

```python
sns.boxplot(x='cty', y='drv', data=df, order=['front', '4-wheel', 'rear']);
plt.ylabel('');
plt.xlabel('city miles per gallon');
plt.title('miles per gallon by driving system');
```

```python
df
```

```python
sns.boxplot(x='cty', y='drv', data=df,
            hue='year',
            order=['front', '4-wheel', 'rear']);
plt.ylabel('');
plt.xlabel('city miles per gallon');
plt.title('miles per gallon by driving system');
```

```python
plt.figure(figsize=(12, 4))
sns.boxplot(x='cty', y='drv', data=df,
            hue='year',
            order=['front', '4-wheel', 'rear']);
plt.ylabel('');
plt.xlabel('city miles per gallon');
plt.title('miles per gallon by driving system');
```

```python
plt.figure(figsize=(12, 4))
sns.boxplot(x='cty', y='drv', data=df,
            hue='year',
            palette='Set2',
            order=['front', '4-wheel', 'rear']);
plt.ylabel('');
plt.xlabel('city miles per gallon');
plt.title('miles per gallon by driving system');
```

```python
plt.figure(figsize=(12, 4))
sns.boxplot(x='cty', y='drv', data=df,
            hue='year',
            palette=['aqua', 'salmon'],
            order=['front', '4-wheel', 'rear']);
plt.ylabel('');
plt.xlabel('city miles per gallon');
plt.title('miles per gallon by driving system');
```

```python
sns.factorplot(y='drv', x='cty', data=df,
              kind='box',
              row='year')
```

```python
sns.catplot(y='drv', x='cty', data=df,
              kind='box',
              col='year')
```

```python
df['gear'] = df.trans.str[0].map({'a':'auto', 'm':'manual'}).astype('category')
df
```

```python
sns.catplot(y='drv', x='cty', data=df,
              kind='box',
              row='year',
              col='gear');
```
