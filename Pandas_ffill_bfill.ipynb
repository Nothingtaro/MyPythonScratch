## fillna: ffill (forward fill), bfill (backward fill) 

```python
import pandas as pd
print(f'pandas version: {pd.__version__}')

%config InlineBackend.figure_format = 'retina'
pd.set_option('display.max_rows', None)
```

```python
do = pd.read_csv('https://datahub.io/core/oil-prices/r/brent-daily.csv', header=0
                , names=['Date', 'Oil'], parse_dates=['Date'])
do
```

```python
do.info()
```

```python
doil = do[(do['Date'] >= '2020-01-01') & (do['Date'] <= '2020-05-31')]
doil
```

## daily SET index 

```python
from_dt = '20200101'
to_dt = '20200531'
#to_dt = pd.Timestamp.now().strftime('%Y%m%d')
url = f'https://stooq.com/q/d/l/?s=^set&d1={from_dt}&d2={to_dt}&i=d'
dset = pd.read_csv(url, parse_dates = ['Date'], usecols = ['Date', 'Close'])
dset
```

```python
dset.columns = ['Date', 'SET_index']
dset
```

```python
df = pd.merge(doil, dset, left_on='Date', right_on='Date', how='outer', indicator=True)
df
```

```python
df = df.sort_values('Date')
df
```

```python
df['_merge'].value_counts()
```

```python
df[df['_merge'] != 'both']
```

```python
df['Oil_f'] = df.Oil.ffill() # forward fill
df['Oil_b'] = df.Oil.bfill() # backward fill
df['SET_f'] = df['SET_index'].ffill()
df['SET_b'] = df['SET_index'].bfill()
df
```

```python
df[['Oil', 'SET_index']].corr()
```

```python
df[['Oil_f', 'SET_f']].corr()
```

```python
df[['Oil_b', 'SET_b']].corr()
```
