```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.head()
```

```python
sns.distplot(df['cty'])
```

```python
sns.distplot(df['cty'], kde=False)
```

```python
sns.displot(df['cty'])
```

```python
sns.displot(df['cty'], kde=True)
```

```python
sns.displot(df['cty']);
plt.axvline(x=df.cty.mean(), color='orange', linestyle='--');
plt.title(f'mean = {df.cty.mean():.2f}, sd = {df.cty.std():.2f}');
```

```python
sns.distplot(df['cty'], kde=False, label="city");
sns.distplot(df['hwy'], kde=False, label="highway");
plt.legend();
```

```python
sns.displot(df['cty']);
sns.displot(df['hwy']);
```

```python
sns.distplot(df['cty'], kde=False, label="city");
plt.show();
sns.distplot(df['hwy'], kde=False, label="highway");
plt.show();
```

```python
fig, ax = plt.subplots(1,2)
print(ax.shape)
```

```python
fig, ax = plt.subplots(1,2, sharex=True, sharey=True)
sns.distplot(df['cty'], kde=False, label="city", ax=ax[0]);
sns.distplot(df['hwy'], kde=False, label="highway", ax=ax[1]);
```
