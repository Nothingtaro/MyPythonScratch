```python
import pandas as pd
%matplotlib inline

pd.set_option('display.max_rows', None)
```

## Read a single sheet 

```python
colnames = ['brand', '1300cc', '1600cc', '1800cc', '2000cc', 'gt2000cc', 'elec', 'total']
df = pd.read_excel('Car1_whole2559.xlsx', sheet_name=1, usecols='B:I',
                  skiprows=7, skipfooter=3,
                  header=None, names=colnames)
df
```

## Read multiple sheets 

```python
colnames = ['brand', '1300cc', '1600cc', '1800cc', '2000cc', 'gt2000cc', 'elec', 'total']
dfs = [] # list for dataframes
dfs = [pd.read_excel('Car1_whole2559.xlsx', sheet_name=sn, usecols='B:I',
                  skiprows=7, skipfooter=3,
                  header=None, names=colnames)
            for sn in range(1, 13)]
```

```python
dfs[0]
```

```python
dfs[11]
```

```python
[d.set_index('brand', inplace=True) for d in dfs]
```

```python
dfs[0]
```

```python
df = pd.concat(dfs, keys=range(1, 13), names=['month', 'brand'])
df.sample(10)
```

```python
df.head(100)
```

```python
df = df[df.total != 0]
```

```python
df.sample(n=10)
```

```python
df.reset_index(inplace=True)
df.head(100)
```

## Save to CSV 

```python
df.to_csv('car1_2559.csv', index=False)
```

## Save multiple dataframes to Excel 

```python
sheets = ['Jan', "Feb", 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
excel_file = pd.ExcelWriter('msheet2559.xlsx')
[c.to_excel(excel_file, sheet_name = sheets[i]) for i, c in enumerate(dfs)]
excel_file.close()
```
